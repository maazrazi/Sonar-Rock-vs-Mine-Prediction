## *Import the Libraries*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## *Get the Data*

```python
df=pd.read_csv(r'C:\Users\Hp\Desktop\Projects\Copy of sonar data.csv',header=None)
```

```python
df.tail()
```

```python
df.describe()
```

```python
df[60].value_counts()
```

```python
df[60] = df[60].map({'R': 1, 'M': 0})
```

```python
df.tail()
```

```python
X=df.drop(columns=60,axis=1)
```

```python
y=df[60]
```

## *Train-Test Split*

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=1)
```

## *Feature Scaling*

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## *Logistic Classifier*

```python
from sklearn.linear_model import LogisticRegression
model_logistic=LogisticRegression()
model_logistic.fit(X_train,y_train)

log_pred = model_logistic.predict(X_test)
```

## *SVM Classifier*

```python
from sklearn.svm import SVC
model_SVC = SVC()
model_SVC.fit(X_train, y_train)

SVM_pred = model_SVC.predict(X_test)
```

## *Random Forest Classifier*

```python
from sklearn.ensemble import RandomForestClassifier
model_rfc = RandomForestClassifier(n_estimators=300)
model_rfc.fit(X_train, y_train)

rfc_pred = model_rfc.predict(X_test)
```

## *Results Comparison*

```python
from sklearn.metrics import classification_report, confusion_matrix
print('\033[1m'+'Logistic Classifier\n'+'\033[0m')
print('Classification report\n',classification_report(y_test,log_pred))
print('Confusion_matrix\n', confusion_matrix(y_test,log_pred))

print('\033[1m'+'\n\nSVM Classifier\n'+'\033[0m')
print('Classification report\n',classification_report(y_test,SVM_pred))
print('Confusion_matrix\n', confusion_matrix(y_test,SVM_pred))

print('\033[1m'+'\n\nRandom Forest Classifier\n'+'\033[0m')
print(' Classification report\n',classification_report(y_test,rfc_pred))
print('Confusion_matrix\n', confusion_matrix(y_test,rfc_pred))
```

## *Plot ROC and compare AUC*

```python
from sklearn.metrics import roc_curve, auc

log_fpr, log_tpr,_ = roc_curve(y_test, log_pred)
auc_log = auc(log_fpr, log_tpr)

svm_fpr, svm_tpr,_ = roc_curve(y_test, SVM_pred)
auc_svm = auc(svm_fpr, svm_tpr)

rfc_fpr, rfc_tpr,_= roc_curve(y_test,rfc_pred)
auc_rfc = auc(rfc_fpr, rfc_tpr)

plt.figure(figsize=(5, 5), dpi=100)
plt.plot(log_fpr, log_tpr, marker='.', label='Log (auc = %0.3f)' % auc_log)
plt.plot(svm_fpr, svm_tpr, linestyle='-', label='SVM (auc = %0.3f)' % auc_svm)
plt.plot(rfc_fpr, rfc_tpr, linestyle='--', label='rfc (auc = %0.3f)' % auc_rfc)


plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

plt.legend()

plt.show()
```
